```
from matplotlib import pyplot
from mpl_toolkits import mplot3d
import numpy as np
```

```
def objective(x, y):
  """ optimize etmemiz gereken fonksyon: """
  return (4 - 2.1*(x**2) + (x**4)/3)*(x**2) + x*y + (4*(y**2) - 4)*(y**2)
```

```
np.random.seed(5)
# Başlatma
min = -2
max = 2
step_size = 0.1
current_x = min + (max - min) * np.random.rand()
current_y = min + (max - min) * np.random.rand()
# başlangıç ​​noktasını değerlendirmesi
solution_eval = objective(current_x, current_y)

scores = list()
scores.append(solution_eval)
t = 100
for i in range(1000):
	# Yeni yer bulması 
  candidate_x = current_x + np.random.randn() * step_size
  candidate_y = current_y + np.random.randn() * step_size
	
  candidate_eval = objective(candidate_x, candidate_y)

  if candidate_eval <= solution_eval:
    current_x  = candidate_x
    current_y = candidate_y
    solution_eval = candidate_eval
    scores.append(solution_eval)
    print('>%d f(%s, %s) = %.5f' % (i, current_x, current_y, solution_eval))
  else:
    p = 1/(1 + np.exp(np.abs(solution_eval - candidate_eval)/t))
    if np.random.rand() < p:
      current_x  = candidate_x
      current_y = candidate_y
      solution_eval = candidate_eval
      scores.append(solution_eval)
      print('>%d f(%s, %s) = %.5f' % (i, current_x, current_y, solution_eval))
  ## uncomment the following for seamulated anealing algorithm 
  # t = t*0.8
print('f(%s, %s) = %f' % (current_x, current_y, solution_eval))
pyplot.plot(scores, '.-')
pyplot.show()
```
